```python
def get_digits(num_grps, prev_divisor=0, remainder=0):
    ans_list = []
    for num in num_grps:
        new_num = remainder * 100 + num
        prev_q = 0
        prev_d = 0
        for r in range(0, 10):
            next_divisor = (prev_divisor + prev_divisor % 10) * 10 + r
            q = next_divisor * r
            if new_num < q:
                ans_list.append(r-1)
                remainder = new_num - prev_q
                next_divisor = prev_d
                break
            elif r == 9:
                ans_list.append(r)
                remainder = new_num - q
                break
            prev_q = q
            prev_d = next_divisor
        prev_divisor = next_divisor
    return prev_divisor, remainder, ans_list

def sqrt_p(n: int, prec: int):
    rev_str = str(n)[::-1]
    num_grps = [int((rev_str[d:d + 2])[::-1]) for d in range(0, len(rev_str), 2)][::-1]
    prec_grps = [0 for d in range(prec)]
    prev_div, remainder, ans_list = get_digits(num_grps)
    t, t1, ans_list1 = get_digits(prec_grps, prev_div, remainder)
    ans_list.append(".")
    ans_list.extend(ans_list1)
    return "".join([str(s) for s in ans_list])
    #print(num_grps, prec_grps)
```
